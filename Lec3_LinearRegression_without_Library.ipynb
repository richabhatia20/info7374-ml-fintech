```python
import pandas as pd
import numpy as np
import csv
```

```python
amazon_data = pd.read_csv('E:\\Information Systems\\FinanceML\\Data\\Equity_Group\\AMZN.csv')
```

```python
amazon_data['price_avg'] = (amazon_data['Open'] + amazon_data['Close'])/2
```

```python
def stock_returns(price_avg):
    st_re = []
    maxi = len(price_avg)
    for index,i in enumerate(price_avg):
        if index == 0:
            st_re.append("NAN")
        if index != 0:
            a = np.log(price_avg[index] / price_avg[index-1])
            st_re.append(a)
    return st_re
```

```python
dist = stock_returns(amazon_data['price_avg'])
```

```python
amazon_data['stock_returns'] = dist
```

```python
yt = amazon_data['stock_returns'] - amazon_data['RF'] 
```

```python
import numpy
np.asarray(yt)
yt = yt.reshape(564,1)
yt.shape
```

```python
yt = np.asarray(yt)
yt.shape
```

```python
xtt = np.matrix(np.asarray([np.asarray(amazon_data['MktRF']), np.asarray(amazon_data['SMB']), np.asarray(amazon_data['HML'])]))
```

```python
xtt.shape
```

```python
xt = np.matrix(np.transpose(xtt))
```

```python
xt.shape
```

```python
mult = xtt*xt
mult.shape
```

```python
beta = np.matrix(((np.linalg.inv(mult))*xtt)*yt)
```

```python
beta.shape
```

```python
beta
```

```python
ypred = np.matrix(xt*beta)
```

```python
ypred
```

```python
residual = yt - ypred
residual.shape
```

```python

%matplotlib inline
import matplotlib.pyplot as plt
plt.plot(ypred, color='red')
plt.plot(yt, color = 'green')
plt.axis([0, 1, 0, -0.2])
plt.show()
```

```python
sigma = np.std(residual)
```

```python
var_beta_hat = np.diagonal((sigma**2)*np.linalg.inv(mult))
var_beta_hat = var_beta_hat.reshape(3,1)
var_beta_hat.shape
```

```python
t = beta/var_beta_hat
t
```
